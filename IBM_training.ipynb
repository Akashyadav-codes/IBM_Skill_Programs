```python
df=df.drop_duplicates()
df
```

```python
#Ensure numeric column have a correct datatype
numeric_col= ['Month_Number', 'Facecream', 'Facewash', 'Toothpaste', 'Bathingsoap',
       'Shampoo', 'Moisturizer', 'Total_Units', 'Total_Profit']
for col in numeric_col:
    df[col]=pd.to_numeric(df[col],errors='coerce')
df
```

```python
#Filling any numeric value with median
for col in numeric_col:
    df[col].fillna(df[col].median(),inplace=True)
    df
```

```python
Groupby: groupby() ka use hum data ko groups me divide karke analysis karne ke liye karte hain
 bilkul SQL ke GROUP BY jaisa

...Basic idea
1.Split  Apply  Combine
2.Split data into groups
3.Apply function (sum, max, mean)
```

```python
#sales trades over the month 
import matplotlib.pyplot as plt

monthly_unit=df.groupby("Month_Number")["Total_Units"].sum()
print("Sales trade over the month:")
print(monthly_unit)
monthly_profit=df.groupby("Month_Number")["Total_Profit"].sum()
print("Sales trade over the month:")
print(monthly_profit)




```

```python
plt.figure(figsize=(10,5))
plt.plot(monthly_unit.index,monthly_unit.values,marker="*",color="red")
plt.title("total unit sold per month ")
plt.xlabel(" Month Number")
plt.ylabel("unit sold")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(monthly_profit.index,monthly_profit.values,marker="o",color="red")
plt.title("total unit sold per month ")
plt.xlabel(" Month Number")
plt.ylabel("unit sold")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(monthly_profit.index,monthly_profit.values)
plt.title("total unit sold per month ")
plt.xlabel(" Month Number")
plt.ylabel("unit sold")
# plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,5))

# Column 1
plt.subplot(1, 2, 1)
plt.plot(monthly_unit.index, monthly_unit.values, marker="*")
plt.title("Monthly Units Sold")
plt.xlabel("Month")
plt.ylabel("Units")
plt.grid(True)

# Column 2
plt.subplot(1, 2, 2)
plt.plot(monthly_profit.index, monthly_profit.values, marker="o")
plt.title("Monthly Profit")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(
    monthly_unit.index,
    monthly_unit.values,
    marker="*",
    color="red",
    label="Monthly Units Sold"
)

plt.plot(
    monthly_profit.index,
    monthly_profit.values,
    marker="o",
    color="blue",
    label="Monthly Profit"
)

plt.title("Monthly Units Sold vs Monthly Profit")
plt.xlabel("Month Number")
plt.ylabel("Value")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd
df=pd.read_csv(r"https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
df
```

```python
df.head()
df.tail()
df.columns
```

```python
numeric_col= ['total_bill', 'tip', 'size']
for col in numeric_col:
    df[col]=pd.to_numeric(df[col],errors='coerce')
df
```

```python
# if you want to see total rows
# pd.set_option("display.max_rows",None)
df
```

```python
revenue_day=df.groupby("day")["total_bill"].sum()
print(revenue_day)
revenue_time=df.groupby("time")["total_bill"].sum()
print(revenue_time)
revenue_gender=df.groupby("sex")["total_bill"].sum()
print(revenue_gender)
```

```python
revenue_day=df.groupby("day")["total_bill"].mean().sort_values(ascending=False)

print(revenue_day)
```
